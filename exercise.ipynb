## Setup

For this exercise, we will use already acquired and postprocessed data. The provided .csv file contains monthly climate data for Marrakesh and Munich for the last 10 years. For each month, it reports:

- the mean atmospheric pressure,
- mean and standard deviation of the daily temperature,
- total monthly precipitation,
- total hours of sunshine, and
- number of rainy days.

Source: Deutscher Wetterdienst

```python
# Load data and print first 10 lines of table
df = pd.read_csv('climate_data.csv')

print(df[0:10].to_string())
```

## Matplotlib and seaborn

The python libraries [matplotlib](https://matplotlib.org/) or [seaborn](https://seaborn.pydata.org/) allow quick and easy plotting of complex data. Many of you will already be familiar with these libraries. If not, have a look at the [tutorial of seaborn](https://seaborn.pydata.org/tutorial/introduction.html), which we will be using today. An example for a bar plot that plots the average monthly temperature for both cities side-by-side looks like this:

```python
sns.set_theme(style='whitegrid')

plt.figure(figsize=(6,6))
g = sns.barplot(data=df, x='Month', y='Monthly temperature - mean [C]', hue='City')
_ = g.set(xticklabels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```
